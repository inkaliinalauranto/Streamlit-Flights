```python
import pandas as pd
import streamlit as st
```

```python
flights_df = pd.read_csv("https://pxdata.stat.fi/PxWeb/sq/a74f5130-03b1-426d-94b3-158f4e262286", encoding="latin1")
flights_df.head()
```

```python
seasonal_flights_df = flights_df.copy()
```

```python
month_columns = [name for name in seasonal_flights_df.columns if "2024M" in name]
month_columns
```

```python
seasonal_flights_df["Yhteensä"] = seasonal_flights_df[month_columns].sum(axis=1)
seasonal_flights_df
```

```python
def get_seasonal_travel_percent(row, month1, month2, month3):
    month1_i = list(row.index).index(f"2024M{month1}")
    month2_i = list(row.index).index(f"2024M{month2}")
    month3_i = list(row.index).index(f"2024M{month3}")
    year_total_i = list(row.index).index("Yhteensä")

    season_total = row.values[month1_i] + row.values[month2_i] + row.values[month3_i]
    year_total = row.values[year_total_i]
    
    return round(season_total / year_total * 100)

seasonal_flights_df["Talvi 2024"] = seasonal_flights_df.apply(lambda row: get_seasonal_travel_percent(row, "12", "01", "02"), axis=1)
seasonal_flights_df["Kevät 2024"] = seasonal_flights_df.apply(lambda row: get_seasonal_travel_percent(row, "03", "04", "05"), axis=1)
seasonal_flights_df["Kesä 2024"] = seasonal_flights_df.apply(lambda row: get_seasonal_travel_percent(row, "06", "07", "08"), axis=1)
seasonal_flights_df["Syksy 2024"] = seasonal_flights_df.apply(lambda row: get_seasonal_travel_percent(row, "09", "10", "11"), axis=1)

seasonal_flights_df= seasonal_flights_df.filter(items=["Ilmoittava lentoasema", 
                                                        "Talvi 2024", 
                                                        "Kevät 2024", 
                                                        "Kesä 2024", 
                                                        "Syksy 2024"], axis=1)

seasonal_flights_df.head()
```
